**# 3. Identify categorical vs Continous variables in our car sales database**

```python
print("\n" + "═" * 60)
print("VARIABLE TYPES")
print("═" * 60)

categorical = [
    'customer_name', 'customer_phone', 'make', 'model', 'color',
    'condition', 'salesperson', 'payment_method', 'trade_in',
    'warranty_purchased', 'date'
]

continuous = [
    'sale_id', 'year', 'mileage', 'sale_price', 'cost_price', 'trade_in_value'
]

print(f"\nCategorical variables ({len(categorical)}):")
for v in categorical:
    print(f"  • {v}")

print(f"\nContinuous / Numeric variables ({len(continuous)}):")
for v in continuous:
    print(f"  • {v}")
```

**# 4. Fix any missing, null, or N/A values under customer_name and salesperson**

*# Doing this because these 2 columns have tons of blanks, NULL values, and N/A values scattered around*

```python
print("\n" + "═" * 60)
print("FIXING MISSING VALUES — customer_name & salesperson")
print("═" * 60)

# 4a. Replace string placeholders with actual NaN across the whole dataframe
df.replace(['NULL', 'N/A', 'null', 'n/a', 'Na', 'NA', ''], np.nan, inplace=True)

print(f"\nMissing customer_name: {df['customer_name'].isna().sum()}")
print(f"Missing salesperson:   {df['salesperson'].isna().sum()}")

# Fill missing customer names with "Unknown Customer"
df['customer_name'].fillna('Unknown Customer', inplace=True)

# Fill missing salesperson with "Unassigned"
df['salesperson'].fillna('Unassigned', inplace=True)

print("\nAfter filling:")
print(f"Missing customer_name: {df['customer_name'].isna().sum()}")
print(f"Missing salesperson:   {df['salesperson'].isna().sum()}")
```

**# 5. Start Full Data sleanup process along with little fixes**

```python
print("\n" + "═" * 60)
print("FULL DATA CLEANUP")
print("═" * 60)
```

```python
# ── 5a. Standardize column names (lowercase, snake_case) ──
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
print("\n✓ Column names standardized")
print(f"  Columns: {list(df.columns)}")

# ── 5b. Remove duplicate rows ──
dupes_before = df.duplicated().sum()
df.drop_duplicates(inplace=True)
print(f"\n✓ Duplicates removed: {dupes_before}")

# ── 5c. Standardize DATE column ──
# The dates are in many formats: MM/DD/YYYY, YYYY-MM-DD, "Jan 28 2023", etc.
df['date'] = pd.to_datetime(df['date'], format='mixed', dayfirst=False)
print(f"\n✓ Dates parsed — sample: {df['date'].iloc[0]} (dtype: {df['date'].dtype})")

# ── 5d. Standardize TEXT columns (title case / uppercase) ──
df['make'] = df['make'].str.strip().str.title()
df['model'] = df['model'].str.strip().str.title()
df['color'] = df['color'].str.strip().str.title()
df['condition'] = df['condition'].str.strip().str.title()
df['salesperson'] = df['salesperson'].str.strip().str.title()
df['customer_name'] = df['customer_name'].str.strip().str.title()
df['payment_method'] = df['payment_method'].str.strip().str.title()
df['trade_in'] = df['trade_in'].str.strip().str.title()
df['warranty_purchased'] = df['warranty_purchased'].str.strip().str.title()

print("✓ Text columns standardized to Title Case")

# ── 5e. Standardize PAYMENT_METHOD (merge "Finance" and "Financing") ──
df['payment_method'] = df['payment_method'].replace('Financing', 'Finance')
print("✓ Payment method: merged 'Financing' → 'Finance'")

# ── 5f. Standardize PHONE NUMBERS → (XXX) XXX-XXXX ──
def clean_phone(phone):
    if pd.isna(phone):
        return np.nan
    digits = ''.join(filter(str.isdigit, str(phone)))
    if len(digits) == 10:
        return f"({digits[:3]}) {digits[3:6]}-{digits[6:]}"
    return np.nan  # bad phone numbers become NaN

df['customer_phone'] = df['customer_phone'].apply(clean_phone)
print(f"✓ Phone numbers standardized — missing: {df['customer_phone'].isna().sum()}")

# ── 5g. Handle remaining MISSING VALUES ──
print("\n── Remaining missing values per column ──")
missing = df.isna().sum()
print(missing[missing > 0])

# Numeric columns: fill with median (robust to outliers)
for col in ['sale_price', 'cost_price', 'trade_in_value']:
    median_val = df[col].median()
    filled = df[col].isna().sum()
    df[col].fillna(median_val, inplace=True)
    if filled > 0:
        print(f"  → {col}: filled {filled} nulls with median ({median_val})")

# trade_in_value should be 0 when trade_in is "No"
df.loc[df['trade_in'] == 'No', 'trade_in_value'] = 0
print("  → trade_in_value set to 0 where trade_in = 'No'")

# Fill remaining missing categoricals
df['customer_phone'].fillna('Unknown', inplace=True)
df['make'].fillna('Unknown', inplace=True)
df['warranty_purchased'].fillna('Unknown', inplace=True)

print("\n✓ All missing values handled")

# ── 5h. Fix DTYPES ──
df['year'] = df['year'].astype(int)
df['mileage'] = df['mileage'].astype(int)
df['sale_price'] = df['sale_price'].astype(float)
df['cost_price'] = df['cost_price'].astype(float)
df['trade_in_value'] = df['trade_in_value'].astype(float)

print("✓ Data types corrected")
```
